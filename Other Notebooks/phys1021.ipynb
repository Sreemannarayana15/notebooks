## For plotting graphs of psi

```python
def psi(n = int, a=int):
    fig, ax = plt.subplots(nrows=n, figsize=(7, 9))
    x = np.linspace(0, a, 100)
    for i in range(1, n+1):
        psi = []
        for j in x:
            psi.append(math.sqrt(2/a) * math.sin(i*math.pi*j/a))
        ax[i-1].plot(x, psi)
    return fig
```

## For plotting graphs of |psi|^2

```python
def psi2(n = int, a=int):
    fig, ax = plt.subplots(nrows=n, figsize=(7, 9))
    x = np.linspace(0, a, 100)
    for i in range(1, n+1):
        psi = []
        for j in x:
            psi.append((math.sqrt(2/a) * math.sin(i*math.pi*j/a))**2)
        ax[i-1].plot(x, psi)
    return fig
```

## For plotting comparision graph of psi and |psi|^2

```python
def prob(n = int, a=int):
    '''Takes in two arguments (n) and (a)
    n(int) : number of energy levels
    a(int) : lenght of the 1-D box'''
    fig, ax = plt.subplots(nrows=n, ncols=2, figsize=(7, 7))
    x = np.linspace(0, a, 100)
    for i in range(1, n+1):
        psi = []
        for j in x:
            psi.append(math.sqrt(2/a) * math.sin(i*math.pi*j/a))
        ax[i-1, 0].plot(x, psi)
        psi = []
        for j in x:
            psi.append((math.sqrt(2/a) * math.sin(i*math.pi*j/a))**2)
        ax[i-1, 1].plot(x, psi)
    return fig
```
