## Challenge

1. In which events did Jesse Owens win a medal <br>
2. Which country has won the most men's gold medals in singles badminton over the years. Sort the results alphabetically by Athlete's name.<br>
3. Which countries have won the most medals in recent years (from 1984 to 2008)? <br>
4. Display the male gold medal winners for the 100m track and field sprint event over the years. List the results starting with most recent. Show the olympic city, edition, athlete, and the country.

```python
# 1. In which events did Jesse Owens win a medal
res = df[df.Athlete == 'OWENS, Jesse']
print(res.Event.unique())
```

```python
# 2. Which country has won the most men's gold medals in singles badminton over the years. 
#Sort the results alphabetically by Athlete's name.
res = df[(df.Sport == 'Badminton') & (df.Event == 'Singles') & (df.Medal == 'Gold')]
res.sort_values(by=['Athlete'])
```

```python
# 3. Which countries have won the most medals in recent years (from 1984 to 2008)?
res = df[(df.Year >= 1984) & (df.Year <= 2008)]
res.value_counts('Country').head(3)
```

```python
# 4. Display the male gold medal winners for the 100m track and field sprint event over the years. 
# List the results starting with most recent. Show the olympic city, edition, athlete, and the country.
res = df[(df.Sport == 'Athletics') & (df.Gender == 'Men') & (df.Event == '100M')]
res.sort_values(by='Year', ascending = False)[['City', 'Year', 'Athlete', 'Country']]
```

# Stack and unstack

## Stack

```python
ef = df[(df.Year == 2008) & ((df.Event == '100M') | (df.Event == '200M'))]
ef = ef.groupby(['Country', 'Discipline', 'Gender', 'Event']).size()
ef = ef.unstack(['Discipline', 'Event'])
```

```python
ef.stack('Event')
```

## Unstack

```python
ef.unstack('Gender', fill_value = 0)
```

# Final Challenge
### In every Olympics, which US athlete has won the most total number of medals? Include the athlete's discipline.

```python
res = df[df.Country == 'USA']
res = df.groupby(['Year', 'Athlete', 'Medal']).size().unstack(fill_value=0)
res['total'] = res['Gold'] + res['Silver'] + res['Bronze']
res.reset_index(inplace=True)
tu = [group.sort_values('total', ascending= False)[:1] for key, group in res.groupby('Year')]
tu
top = pd.DataFrame()
for i in tu:
    top = top.append(i)
top
```
